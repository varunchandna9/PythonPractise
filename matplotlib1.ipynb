```python
from matplotlib import pyplot as plt
days=[1,2,3,4,5]
sleeping=[7,8,6,11,7]
eating=[2,3,4,3,2]
working=[7,8,7,2,2]
playing=[8,5,7,8,13]
plt.plot([],[],color="m",label="sleeping",linewidth=5)
plt.plot([],[],color="c",label="eating",linewidth=5)
plt.plot([],[],color="r",label="working",linewidth=5)
plt.plot([],[],color="k",label="playing",linewidth=5)

plt.stackplot(days,sleeping,eating,working,playing,                 #area plot used to track changes over time or groups are used track two more related gropus that make one whole category
color=["m","c","r","k"])

plt.xlabel("x")
plt.ylabel("y")
plt.title("stack plot")
plt.legend()
plt.show()



```

```python
import numpy as np
import matplotlib.pyplot as plt                     #working with multiple plot 
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)
t1=np.arange(0.0,5.0,0.1)
t2=np.arange(0.0,5.0,0.02)
plt.subplot(211)                #subplot is used to plot multiple plot 211  2 means two plot... horizontial we have only one plot and vertical we have 1 plot  
plt.plot(t1,f(t1),"bo",t2,f(t2))
plt.subplot(212)
plt.plot( t2,np.cos(2*np.pi*t2))
plt.show()
```
